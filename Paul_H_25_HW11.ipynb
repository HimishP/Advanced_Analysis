3. Write a method that takes another particle as an argument and finds the distance between the two particles

4. Write a method that given a time interval ``dt`` will update the position of the particle to the new position based on the current position and velocity.

5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np
class Particle(object):
    def __init__(self, x, y, z, vx, vy, vz, m):
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.m = m
    def Kinetic_Energy(self):
        return (0.5*(self.m)*(self.vx**2+self.vy**2+self.vz**2))
    def distance(self, obj2):
        return np.sqrt((self.x-obj2.x)**2 + (self.y-obj2.y)**2 + (self.z-obj2.z)**2)
    def new_position(self, dt):
        self.x = self.x + self.vx*dt
        self.y = self.y + self.vy*dt
        self.z = self.z + self.vz*dt
        return(self.x, self.y, self.z)

# part 5
class ChargedParticle(Particle):
    
    def __init__(self, x, y, z, vx, vy, vz, m, charge):
    
        super().__init__( x, y, z, vx, vy, vz, m)
        # Add a new attribute for electric charge
        self.charge = charge
    def Charged_Interaction(self, obj3):
        seperation = self.distance(obj3)
        if seperation < 0.25:
            if self.charge == -obj3.charge:
                self.vx = 0
                self.vy = 0
                self.vz = 0
                obj3.vx = 0
                obj3.vy = 0
                obj3.vz = 0
                self.charge = 0
                obj3.charge = 0
                return(self.vx, self.vy, self.vz, obj3.vx, obj3.vy, obj3.vz)
            else:
                self.vx = self.vx * (-1*(self.charge + obj3.charge)*(self.m/obj3.m))
                self.vy = self.vy * (-1*(self.charge + obj3.charge)*(self.m/obj3.m))
                self.vz = self.vz * (-1*(self.charge + obj3.charge)*(self.m/obj3.m))
                obj3.vx = obj3.vx * (-1*(self.charge + obj3.charge)*(obj3.m/self.m))
                obj3.vy = obj3.vy * (-1*(self.charge + obj3.charge)*(obj3.m/self.m))
                obj3.vz = obj3.vz * (-1*(self.charge + obj3.charge)*(obj3.m/self.m))
                return("repel", self.vx, self.vy, self.vz, obj3.vx, obj3.vy, obj3.vz)
            return("Interaction")

def simulate(p1, p2, dt, steps):
    for step in range(steps):
        p1.new_position(dt)
        p2.new_position(dt)
        p1.Charged_Interaction(p2)
        t = (step + 1) * dt
        print("step", step+1, "t =", t)
        print("  P1: pos=", (p1.x, p1.y, p1.z), "vel=", (p1.vx, p1.vy, p1.vz), "q=", p1.charge)
        print("  P2: pos=", (p2.x, p2.y, p2.z), "vel=", (p2.vx, p2.vy, p2.vz), "q=", p2.charge)
        
p1 = ChargedParticle(-5,-5,-5,1,1,1,4,0.5)
p2 = ChargedParticle(5,5,5,-1,-1,-1,3,-0.5)
p3 = ChargedParticle(-5,-5,-5,1,1,1,4,0.5)
p4 = ChargedParticle(5,5,5,-1,-1,-1,3,2.0)
print("Test 1, merging")
simulate(p1, p2, 0.05, 300)

        
```

```python
print("Test 2, repulsion")
simulate(p3, p4, 0.05, 300)
```
